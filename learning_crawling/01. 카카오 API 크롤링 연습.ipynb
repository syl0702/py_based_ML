```python
KAKAO_API_KEY = "2926350bcb7a5ebd06910768bc87c213"
KAKAO_WEB_SEARCH_URL = "https://dapi.kakao.com/v2/search/web?query={}&page={}&size=50"
```

카카오 api에 요청하기 위해서는 인증 정보가 필요!

```python
header = {
  "Authorization": "KakaoAK {}".format(KAKAO_API_KEY)
}

header
```

```python
import requests
# 초대권(header) 들고 식당 찾아가기
response = requests.get(KAKAO_WEB_SEARCH_URL.format("장원영"), headers=header) # 요청 URL 및 헤더 정보 넣어주기
response
```

```python
# 응답 코드 가져오기
response.status_code
```

응답 코드(Response Code)
* `1xx` : 처리 중 ( 볼 일이 없다 )
* **`2xx`** : Success
* `3xx` : Redirection ( 추가 동작이 필요하다. - 볼 일이 많이 없다 )
* **`4xx`** : Client Error ( 클라이언트에서 오류 발생 )
    - `404` : Page Not Found (잘못된 url로 요청함)
* **`5xx`** : Server Error ( 서버에서 오류발생 )


```python
datas = response.json()
datas
```

# [실습]
첫 페이지가 아닌 모든 페이지의 데이터를 검색해서 데이터 프레임으로 저장하기

```python
all_datas = []
for i in list(range(1, 51)):
  REQ_URL = KAKAO_WEB_SEARCH_URL.format("장원영", i)
  response = requests.get(REQ_URL, headers=header)
  datas = response.json()['documents']
  all_datas += datas

df = pd.DataFrame(all_datas)
df
```

```python
df.to_csv('장원영_검색.csv', encoding='utf=8')
```
