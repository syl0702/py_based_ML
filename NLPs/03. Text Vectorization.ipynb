# 사이킷런을 이용한 텍스트 벡터화

## CountVectorizer
- DTM을 만드는 데 목적이 있다.
- DTM(Document-Term Matrix)란 문서와 관련된 단어들의 빈도 정보를 행렬 형태로 나타내는 것.
  - 행 : 각 문서를 나타내는 행
  - 열: 어휘의 각 단어를 나타냄. 열의 수는 어휘 크기에 따라 달라짐.
  - 셀 : 단어의 출현 빈도수

- DTM을 통해 단어 중요도/텍스트 표현/문서 유사도 파악 가능

```python
from sklearn.feature_extraction.text import CountVectorizer

cnt_vector = CountVectorizer()
cnt_vector.fit(corpus)
```

```python
feature_vector = cnt_vector.transform(corpus)
feature_vector
```

```python
feature_vector.toarray()
```

- cnt_vector.vocabulary_: 정수 인코딩을 위한 단어 집합

```python
# 단어 확인
vocabs = sorted(cnt_vector.vocabulary_.items()) # 단어 순서대로 정렬
vocabs = [ item[0] for item in vocabs ] # 단어만 뽑아내기
vocabs
```

```python
import pandas as pd

dtm = pd.DataFrame(
    columns=vocabs,
    data=feature_vector.toarray()
)
dtm
```

# 20 뉴스그룹 분류

```python
from sklearn.datasets import fetch_20newsgroups

news_data = fetch_20newsgroups(subset='all', random_state=42)
```

```python
# target 이름 확인
news_data.target_names
```

```python
# 데이터 확인
print(news_data['data'][0])
```

```python
#  훈련 데이터 가져오기. header, footers, 특수문자를 제거하고 불러오기
train_news = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'), random_state=42)
X_train = train_news['data']
y_train = train_news['target']

test_news = fetch_20newsgroups(subset='test', remove=('headers', 'footers', 'quotes'), random_state=42)
X_test = test_news['data']
y_test = test_news['target']
```

```python
# train용 corpus 확인
X_train[:3]
```

```python
# Text Feature를 벡터화 시켜보기
#  - CountVectorizer or TfidfVectorizer 를 사용

cnt_vect = CountVectorizer()
X_train_cnt_vect = cnt_vect.fit_transform(X_train)
X_test_cnt_vect  = cnt_vect.transform(X_test)
```

```python
X_train_cnt_vect.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_clf = LogisticRegression(solver='liblinear')

lr_clf.fit(X_train_cnt_vect, y_train)
pred = lr_clf.predict(X_test_cnt_vect)

accuracy_score(y_test, pred)
```

```python
# 예측할 때는 문장 또는 문서를 벡터화 시켜서 넣어야 한다.
predict_news_vect = cnt_vect.transform(["SAN DIEGO — Money was flying and the clock was spinning when Mets General Manager Billy Eppler emerged from his suite late Wednesday afternoon on the final day of the winter meetings to announce that Justin Verlander indeed had signed a two-year, $86.6 million deal with the club."])
predict_news_vect
```

```python
lr_clf.predict(predict_news_vect)
```
